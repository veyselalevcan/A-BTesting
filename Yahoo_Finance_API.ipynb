#Duraganlıgı gormek

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
```

```python
hisseler= ['XU100.IS', 'THYAO.IS', 'EREGL.IS']
veri= yf.download(hisseler, period='10Y', interval='1mo')
veri= veri['Close']['EREGL.IS']
plot_acf(veri)
plt.show()
```
